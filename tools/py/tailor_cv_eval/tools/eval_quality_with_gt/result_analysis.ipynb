# Evaluation Results Analysis
Analyze and display Promptfoo CV tailoring prompt evaluation results.

```python
import re
import sys
from pathlib import Path

import pandas as pd
from IPython.display import display  # noqa: A004

# Add tools/py to path to import helpers
sys.path.append(str(Path.cwd().parents[2]))
from helpers.promptfoo_helper import PromptfooCsvCols

# Import globals from run.py
sys.path.append(str(Path.cwd()))
from run import RESULTS_CSV

pd.set_option("max_colwidth", 120)

df = pd.read_csv(RESULTS_CSV)

for col in [PromptfooCsvCols.PROVIDER_ID, PromptfooCsvCols.PROVIDER_LABEL]:
    display(df[col].drop_duplicates().sort_values().head().reset_index(drop=True))
```

```python
df = df.sort_values(
    by=[PromptfooCsvCols.SCORE, PromptfooCsvCols.COMPLETION_TOKENS, PromptfooCsvCols.LATENCY],
    ascending=[False, False, True],
)

# fix long prompt label - leave only file name
df[PromptfooCsvCols.PROMPT_LABEL] = df[PromptfooCsvCols.PROMPT_LABEL].apply(lambda x: re.findall(r"\w+\.md", x)[0])

df[
    [
        PromptfooCsvCols.PROVIDER_LABEL,
        PromptfooCsvCols.PROMPT_LABEL,
        PromptfooCsvCols.SUCCESS,
        PromptfooCsvCols.SCORE,
        PromptfooCsvCols.PASSED,
        PromptfooCsvCols.FAILED,
        PromptfooCsvCols.LATENCY,
        PromptfooCsvCols.COMPLETION_TOKENS,
        PromptfooCsvCols.FAILURE_REASON,
    ]
].head(20).reset_index(drop=True)
```
